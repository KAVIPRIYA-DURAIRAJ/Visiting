```python
import pandas as pd
```

```python
placedf = pd.read_csv('Top Indian Places to Visit.csv')
placedf
```

```python
placedf.info()
```

```python
placedf = placedf.drop(columns=['Zone','State','City','Name','Establishment Year','time needed to visit in hrs','Airport with 50km Radius','Weekly Off','Significance','DSLR Allowed','Best Time to visit'])
placedf
```

```python
placedf = placedf.dropna()
placedf
```

```python
placedf.isna().sum()
```

```python
placedf = pd.get_dummies(placedf,dtype=int)
placedf
```

```python
X = placedf.drop(columns=['Entrance Fee in INR'])
Y = placedf['Entrance Fee in INR']
X
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train,Y_train)
```

```python
linear_model.fit(X_train,Y_train)
```

```python
linear_model.score(X_train,Y_train)
```

```python
linear_model.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso
linear_model2 = Lasso(alpha=10,max_iter=70,tol=0.1)
linear_model2.fit(X_train,Y_train)
```

```python
linear_model2.score(X_train,Y_train)
```

```python
linear_model2.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
linear_model3 = Ridge(alpha=25,max_iter=100,tol=0.1)
linear_model3.fit(X_train,Y_train)
```

```python
linear_model3.score(X_train,Y_train)
```

```python
linear_model3.score(X_test,Y_test)
```
